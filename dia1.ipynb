Escribir una funcion que nos pregunte las caracteristicas de un elemento a comprar e imprima el elemento con sus caracteristicas.

```python
#definir funcion y parametros
def elemento_solicitado(caracteristica_1, caracteristica_2):
    print(f"El usuario solicita un telefono de la marca {caracteristica_1} de color {caracteristica_2}")

#Llamar a funcion con argumentos
elemento_solicitado(input("Marca de telefono?"),input("Color del telefono?"))
```
